```python
#preprocessing on a dataset-1
import pandas as pd

import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.replace("-", np.nan, inplace=True)
print(df)
```

```python
# step 2 : replace '-' with NaN to mark missing values

df.replace('-', np.nan, inplace=True)

# convert columns to numeric where needed

df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Marks"] = pd.to_numeric(df["Marks"], errors='coerce')

print(df)
```

```python
# step 3: check missing values

print("Missing Values")
print(df.isnull().sum())
```

```python
# Fill the Age with mean

df["Age"].fillna(df["Age"].mean(), inplace= True)
print(df)
```

```python
# fill the marks with median

df["Marks"].fillna(df["Marks"].median(), inplace=True)
print(df)
```

```python
# step 4: Fix inconsistent text formatting

df["Department"] = df["Department"].str.upper()
print(df)
```

```python
# step 5: convert admission date to standard datetime format

df["Admission_Date"] = pd.to_datetime(df["Admission_Date"],dayfirst=True, errors='coerce')
print(df)
```
